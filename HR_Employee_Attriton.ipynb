```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, classification_report

# Assuming 'df' is your DataFrame containing the employee attrition data

# Define features (X) and target variable (Y)
X = df.drop('Attrition', axis=1)
Y = df['Attrition']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)

# Preprocessing pipeline for handling categorical variables
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numerical_cols),
        ('cat', LabelEncoder(), categorical_cols)
    ], remainder='passthrough'
)

# Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)

# Create a pipeline with preprocessing and classifier
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', rf_classifier)])

# Fit the model
clf.fit(X_train, Y_train)

# Predict on the test set
Y_pred = clf.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)
print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
# Confusion matrix visualization
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(Y_test, Y_pred)
class_names = [0, 1]
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, cmap='Purples', fmt='g', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Logistic Regression setup
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)

# Get accuracy
accuracy = log_reg.score(X_test, Y_test)
print("Model testing accuracy:", accuracy)

# Show confusion matrix and classification report for the model on the test data
Y_pred = log_reg.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:\n", cm)

class_names = ['Not Attrition', 'Attrition']
print("Classification Report:\n", classification_report(Y_test, Y_pred, target_names=class_names))

```

```python
# Confusion matrix visualization
fig, ax = plt.subplots()
tick_marks = range(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
sns.heatmap(pd.DataFrame(cm), annot=True, cmap='Greens', fmt='g')
ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.title('Confusion matrix for Logistic Regression')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# SVM setup
svm_model = SVC()
svm_model.fit(X_train, Y_train)

# Get accuracy
accuracy = svm_model.score(X_test, Y_test)
print("Model testing accuracy:", accuracy)

# Show confusion matrix and classification report for the model on the test data
Y_pred = svm_model.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:\n", cm)

class_names = ['Not Attrition', 'Attrition']
print("Classification Report:\n", classification_report(Y_test, Y_pred, target_names=class_names))


```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# KNN Classifier setup
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)

# Get accuracy
accuracy = knn.score(X_test, Y_test)
print("Model testing accuracy:", accuracy)

# Show confusion matrix and classification report for the model on the test data
Y_pred = knn.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:\n", cm)

class_names = ['Not Attrition', 'Attrition']
print("Classification Report:\n", classification_report(Y_test, Y_pred, target_names=class_names))


```

```python
# Confusion matrix visualization
fig, ax = plt.subplots()
tick_marks = range(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
sns.heatmap(pd.DataFrame(cm), annot=True, cmap='Reds', fmt='g')
ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.title('Confusion matrix for K-Nearest Neighbors')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
# Instantiate classifiers
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
log_reg = LogisticRegression()
svm_model = SVC()
knn = KNeighborsClassifier(n_neighbors=5)

# Train classifiers
rf_classifier.fit(X_train, Y_train)
log_reg.fit(X_train, Y_train)
svm_model.fit(X_train, Y_train)
knn.fit(X_train, Y_train)

# Calculate accuracy for each classifier
accuracy_rf = rf_classifier.score(X_test, Y_test)
accuracy_logreg = log_reg.score(X_test, Y_test)
accuracy_svm = svm_model.score(X_test, Y_test)
accuracy_knn = knn.score(X_test, Y_test)

plt.rcdefaults()
fig, ax = plt.subplots()
algorithms = ('Logistic Regression', 'K Neighbors Classifier', 'Random Forest Classifier','Support Vector Machine')
y_pos = np.arange(len(algorithms))
accuracies = [accuracy_logreg, accuracy_knn, accuracy_rf, accuracy_svm]
ax.barh(y_pos, accuracies, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(algorithms)
ax.invert_yaxis()
ax.set_xlabel('Performance')
ax.set_title('Which one is the best algorithm?')
for i, v in enumerate(accuracies):
    ax.text(v + 1, i, str(v), color='black', va='center', fontweight='normal')
plt.show()
```
