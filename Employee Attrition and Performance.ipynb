```python
sns.boxplot(x='Attrition',y='DistanceFromHome',data=data)
plt.title('Employee Attrition Vs Distance')
plt.show()
```

```python
sns.countplot(x='JobRole', hue='Attrition',data=data)
plt.title("Attrition by Job Role")

plt.xticks(rotation=90)
plt.show()
```
